```python
results = pd.DataFrame()
# fund = filter_funds['Ticker'].iloc[0:457]
fund = filter_funds['Ticker'].iloc[0:2]
for f in fund:
    data = yf.download(tickers=f, period='max', interval='3mo', group_by='ticker')
    data['Ticker'] = f
#     data.head()
    results = results.append(data)
#     rSeries = pd.Series(results)

merge = filter_funds.merge(results,on='Ticker',how='outer')
merge
# merge[merge['Ticker'] == 'FTIWX']
```

```python
merge.dropna(how='any', inplace=True)
merge
```

```python
merge['Prev Close'] = merge['Close'].shift(1)
merge['Prev Adj Close'] = merge['Adj Close'].shift(1)
merge.head(4)
```
